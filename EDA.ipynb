<h2><b><i> ANALYSIS OF PROFIT AND SALES

<h2>Segment wise  dealings of Profit and Sales

```python
dt_seg=data.groupby(['Segment'])[['Profit','Sales']].mean()
print(dt_seg)
y=  sns.catplot(y="Sales",
               x="Segment",
                data=data,
                kind='bar')
x= sns.catplot(y="Profit",
               x="Segment",
                data=data,
                kind='bar')
print(x,y)
```

```python
#analysis of categories

x=  sns.catplot(x="Sales",
               y="Sub-Category",
                data=data,col='Category',
                kind='bar')
y= sns.catplot(x="Profit",
               y="Sub-Category",
               col='Category',
                data=data,
                kind='bar')
```

```python
dt_seg=data.groupby(['Segment'])[['Profit','Sales']].mean()
dt_seg.head()
```

```python
dt_seg.plot(kind='bar',
             width=0.2,
            title='COMPARISION OF PROFIT AND SALES OF SEGMENTS',
            fontsize=10,
                   figsize= (10,8))
```

<h4>From the above figure we see that the 'Sales' of the three segments are pretty good but the 'Profit' is not as high as 'Sales' and can be improved.

```python
dt_reg=data.groupby(['Region'])[['Sales','Profit']].mean()
dt_reg.head()
```

```python
dt_reg.plot(kind='bar',width=0.4,title='Comparision of REGIONS based on Profit and Sales',figsize=(10,8),colormap='viridis')
```

<h4>We can also understand the above code percentage wise in a pie chart<br>
<h2>REGION WISE ANALYSIS USING PIE CHARTS

```python
dt_reg.plot.pie(subplots=True,
                autopct='%1.1f%%',
               figsize=(12,8),
               explode=[0,0,0.1,0.1])
```

<h2> Analysis of Profit and Sales
<h3><li> Sales in South are highest (sales of all are nearly similar )
 <li> Profit in West is the highest .
<li> In South region Profit can be improved significantly.
<li> In  East region profit is 2nd highest and can be improved.
<li> In Central region profit is the least and should be improved. </h3>

<h1><u>STATE WISE ANALYSIS OF PROFIT AND SALES

```python
dt_st=data.groupby(['State'])[['Profit','Sales']].mean()
dt_st.head()
```

```python
dt_st['Sales'].plot.pie(autopct='%1.2f%%',figsize=(22,15),  label='State', colormap='prism')
plt.title('State wise analysis of Sales', fontsize=25)
plt.title('Sub-Category Wise Sales Analysis', fontsize = 20)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

<p> <li>Wyoming has the highest Sales <br>
 <li>It is followed by Vermont </p>

```python
dt_st=dt_st.sort_values('Profit')
dt_st['Profit'].head()
```

```python
data['State'].nunique()
```

```python
dt_st['Profit'].plot(kind='bar',figsize=(20,17), colormap='hot')
plt.title('State-wise Analysis of Profit', fontsize=21)
```

    <li>10 states have Negative Profit
     <li>Ohio has least Profit
